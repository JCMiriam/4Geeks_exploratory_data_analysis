Empezamos limpiando los datos, unificando un único tipo por columna para poder realizar operaciones y trabajar con los datos correctamente.

```python
# Get each column type and check if there are more than one type per column to unify them
column_types = df.dtypes

formatted_types = {
    column: column_types[column].name if column_types[column] != 'bool' else 'boolean' 
    for column in df.columns
}

def get_types_per_column(df_columns):
    for column in df_columns:
        column_types = df[column].apply(type).unique()
        if len(column_types) > 1:
            type_names = [t.__name__ for t in column_types]
            print(f'{column}: <{type_names}>')
        else:
            print(f'{column}: <{column_types[0].__name__}>')

get_types_per_column(df.columns)
```

```python
# Clean data types
df['name'] = df['name'].astype(str)
df['host_name'] = df['host_name'].astype(str)
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

mean_date = df['last_review'].dropna().mean()
df['last_review'] = df['last_review'].fillna(mean_date)
df['last_review'] = df['last_review'].dt.date

mean_reviews_per_month = df['reviews_per_month'].dropna().mean()
df['reviews_per_month'] = df['reviews_per_month'].fillna(mean_reviews_per_month)

get_types_per_column(df.columns)
```

```python
# Search empty data on columns
null_counts = df.isnull().sum()
print(null_counts)
```

```python
# Get duplicate data and remove it if necessary
def get_duplicates(df, column):
    duplicated = df[df.duplicated(subset=[column], keep=False)]
    return duplicated.groupby(column).size().reset_index(name='count')

def get_duplicated_column_data(df, column, value):
    return df[df[column]] == value

# Check that there's no id or complete content duplicated
duplicated_content =  df.duplicated(subset=df.columns.difference(['id'])).any()
print(f'Duplicated content:\n{(duplicated_content)}\n')

duplicated_id = get_duplicates(df, 'id')
print(f'Duplicated id:\n{duplicated_id}')
```

No se encuentra ningún id duplicado ni fila que tenga todos sus datos duplicados. Sin embargo, se procede a hacer una búsqueda de duplicados por nombre, ya que se ha podido registrar una misma estancia varias veces erróneamente, por lo que sus valores podrían ser diferentes siendo la misma.

```python
# Check if there are duplicated names and in case, check whole content of these rows
def get_filter_rows_data(df, column, value, data):
    filtered_df = df[df[column] == value]

    if data:
        filtered_df = filtered_df[data]
    
    return filtered_df

duplicated_name = get_duplicates(df, 'name')
print(f'Duplicated name:\n{duplicated_name}\n')
```

Como hay muchos casos en los que los datos con nombres duplicados tienen los mismos datos, por lo que se entiende que son estancias que se han registrado múltiples veces. Por ello se procede a analizar los diferentes datos para unificarlos en una sola entrada, obteniendo medias, sumas o tratando cada uno dependiendo del contexto.

```python
# Get complete content of entries with duplicate names
names_list = duplicated_name['name'].drop_duplicates().tolist()

columns_data = [
    'id',
    'name',
    'host_id',
    'host_name',
    'neighbourhood_group',
    'neighbourhood',
    'latitude',
    'longitude',
    'room_type',
    'price',
    'minimum_nights',
    'number_of_reviews',
    'last_review',
    'reviews_per_month',
    'calculated_host_listings_count',
    'availability_365'
]

df_dict = {}

for name in names_list:
    filtered_data = get_filter_rows_data(df, 'name', name, columns_data)
    df_dict[name] = pd.DataFrame(filtered_data)

df_names = {name: df for name, df in df_dict.items()}

# for df in df_names.items():
#     print(df)

first_name, first_df = next(iter(df_names.items()))

df_1 = pd.DataFrame(first_df)
df_1
```

```python
# Unify entries with duplicate names
def unify_duplicate_data(df):
    unified_data = {
        'id': df.iloc[0]['id'],
        'name': df.iloc[0]['name'],
        'host_id': df.iloc[0]['host_id'],
        'host_name': df.iloc[0]['host_name'],
        'neighbourhood_group': df.iloc[0]['neighbourhood_group'],
        'neighbourhood': df.iloc[0]['neighbourhood'],
        'latitude': df.iloc[0]['latitude'],
        'longitude': df.iloc[0]['longitude'],
        'room_type': df.iloc[0]['room_type'],
        'price': df['price'].mean(),
        'minimum_nights': df['minimum_nights'].mean(),
        'number_of_reviews': df['number_of_reviews'].sum(),
        'last_review': df['last_review'].max(),
        'reviews_per_month': df['reviews_per_month'].mean(),
        'calculated_host_listings_count': df['calculated_host_listings_count'].mean(),
        'availability_365': df['availability_365'].mean()
    }
    
    return unified_data

unified_data = []

for name, df in df_names.items():
    unified_entry = unify_duplicate_data(df)
    unified_data.append(unified_entry)

print(unified_data)
```
